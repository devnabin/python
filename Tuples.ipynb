```python
# Data Types 
# Tuple
t1 = (1, (2,3), [4,5])
print(t1[1]) # (2,3)
# We can't assign the value in tuple except list values
# t1[0] = 1 ❌
t1[2][0]= 5 # ✅
print(t1) 
```

```python
# More Tuples Tips
#Packing or unpacking Tuples values
person = ('Nabin b', 26, 'Nepal')
name, age, country = person
print(f'Student name is {name}, He is {age} years old and he is from {country}.')

# Tips
t = (1,2,3,4,5,6,2,2)
print(len(t)) # 8 -> total number of items in tuple
print(t.count(2)) # 3 -> number of occurance of item 2, here 2 is occured in 3 times
print(t.index(4)) # 4 -> this gives the index number of the value 4 i.e 4 is in 3rd index
print(2 in t) # True -> item 2 is found in tuple
print(9 in t) # False -> item 9 is not found in tuple
```
