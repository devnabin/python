```python
def linear_search(arr, target):
    for i in range(len(arr)):
        
        if target == arr[i]:
            print(target)
            return i
                     # return -1 if not found
    return -1
# Example usage
numbers = [10, 20, 30, 40, 50, 30]
print(linear_search(numbers, 30))  # Output: 2
print(linear_search(numbers, 30))  # Output: 2
print(linear_search(numbers, 100)) # Output: -1


```
