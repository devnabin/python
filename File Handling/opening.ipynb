```python
# Writing and creating a file

textMsg = "Hello world!, Its me nabin a software engineer at google!"

# Opening a file -> open('filename', 'mode')
fk = open('Hello.txt', 'w')

# writing a fiel
fk.write(textMsg)
fk.close()
```

```python
f = open('Hello.txt', 'r')
content = f.read()
f.close()

f = open('Hello.txt' , 'w')
f.write(content.replace('google', 'nvedia'))
f.close()
```

```python
with open('Hello.txt', 'r') as f: # with auto close the fiel
    for line in f:
        print(line.strip())
```

```python
# append data
f = open('Hello.txt','a')
f.write('\n Yes i will be data engineer!!')
f.close()
```

```python
import json

person = {
    "person1" : {
        "name" : 'X',
        "age" : 56,
        "hobby" : ['dancing', 'singing']
    },
    "person2" : {
        "name" : 'X',
        "age" : 56,
        "hobby" : ['dancing', 'singing']
    },

    "person3" : {
        "name" : 'X',
        "age" : 56,
        "hobby" : ['dancing', 'singing']
    }
}

# Dumping Json Data
with open('person.json', 'w') as fuuu:
    json.dump(person, fuuu)

# Reading Json Data
with open('person.json', 'r') as fuu:
    jsonData = json.load(fuu)
    print(jsonData['person1'])
```
