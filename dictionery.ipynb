```python
person = {
    'Person' : { 'name' : "sabin", 'age' : 34, 'country' : 'Nepal'},
    'Person1' : { 'name' : "nbn", 'age' : 34, 'country' : 'Nepal'},
    'Person2' : { 'name' : "nbn", 'age' : 34, 'country' : 'Nepal'},
    'Person3' : { 'name' : "nbn", 'age' : 34, 'country' : 'Nepal'},
}

print(person['Person'])

#adding new key
person['Person']['graduationDate'] = 2027

# updating
person['Person']['name'] = 'Nabin'
person['Person']['age'] = 27

# accessing dic in two different ways
print(person['Person']) # normal way
print(person.get('Person')) # using get()
```

```python
# Removing and deleting item from dict

# 1: removing by key value pair
print(person['Person1'].pop('name')) # name : 'nbn' will be remove

# 2: remove last inserted item and remove it
print(person['Person1'].popitem())

# 3: remove by delete key directly
del person['Person1']['age']

# removing all the value from key
person['Person2'].clear()
```
