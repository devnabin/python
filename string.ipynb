```python
# Basic string operation for capitalization
a = "heLlo world"
print(a.upper())
print(a.lower())
print(a.title())
print(a.capitalize())
```

```python
# Slicing on string
print(a[:3]) # ending index is 0 so, output will be heL
print(a[0:5:2]) # ending index is 5 with the skip of 2 item so hLo
print(a[::3]) # starting and ending is 0 with the skip of 3 so, hlwl
```

```python
# Removing white space
aText = '     Hello World!   '
print(aText.strip()) # remove space from both side of string # very useful
print(aText.lstrip()) # remove extra space from left side 
print(aText.rstrip()) # remove extra space from right sife of the string
```

```python
# Finding and replacement test
aParagraph = "Hello there, I love AI"
bPar = aParagraph.replace('AI', 'Py Programming')
print(bPar.find('love')) # It will give the first index number where "love" is started 
print(bPar)
```

```python
# splitting and joining
myText = 'Hello I love python Programming'
anotherText = myText.split(' ')
print(anotherText)
print(' '.join(anotherText))
```

```python
# some tips
tipstxt = 'cs46'
print(tipstxt.isalpha()) # result true on all alphabets no numeric
print(tipstxt.isdigit()) # result ture on numbers no alpha
print(tipstxt.isalnum()) # result true on both condition
```
