Set {}
1. Set are the unordered collection of unique elements
2. No dublicate data are alloweded

```python
# No dublicate element
a = {1,2,3,4,4}
print(a)
```

```python
# Set Operation
a = {1,2,3,4}
b = {4,5,6,7}

print(a|b)
print(a&b)
print(a-b)
print(b-a)
```

```python
a ={1,2,3,4,4}
a.add(5) # it will add value 5
a.update([6,7,8]) # it will add and update new values
a.pop() # it will remove one random value
a.remove(4) # if item is not found on remove then it will throw error
a.discard(5) # if item is not found on discard it will not throw error
print(a)
a.clear() # it will remove all the values from the set
```
